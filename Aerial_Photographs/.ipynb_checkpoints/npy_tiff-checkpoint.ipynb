```python
import os
import rasterio
import numpy as np

# Path to the output folder containing the TIFF images
output_folder = r'C:\Users\User\Documents\Python Scripts\Data\RGB_Test\npy_tiff'

# Loop through each image in the directory
for image_file in os.listdir(output_folder):
    if image_file.lower().endswith('.tif'):  # Check for both upper and lower case extensions
        print(f"\nProcessing image: {image_file}")
        image_file_path = os.path.join(output_folder, image_file)
        
        # Open the image
        with rasterio.open(image_file_path) as dataset:
            # Retrieve and print the metadata
            metadata = dataset.meta
            print("\nImage Metadata:")
            for key, value in metadata.items():
                print(f"{key}: {value}")
            
            # Print additional metadata information
            print("\nAdditional Metadata:")
            print(f"CRS (Coordinate Reference System): {dataset.crs}")
            print(f"Bounds: {dataset.bounds}")
            print(f"Width: {dataset.width}")
            print(f"Height: {dataset.height}")
            print(f"Number of Bands: {dataset.count}")
            
            # Loop through each band and check if they are empty
            for i in range(1, dataset.count + 1):  # Band indices are 1-based in rasterio
                band_data = dataset.read(i)
                if np.all(band_data == 0):
                    print(f"Band {i} is empty (all values are zero).")
                else:
                    print(f"Band {i} is not empty.")
                    print(f"Band {i} summary:")
                    print(f"  Min value: {np.min(band_data)}")
                    print(f"  Max value: {np.max(band_data)}")
                    print(f"  Mean value: {np.mean(band_data)}")
```

```python
import os
import rasterio
import numpy as np

# Path to the output folder containing the TIFF images
output_folder = r'C:\Users\User\Documents\Python Scripts\Data\RGB_Test\npy_tiff'

# Loop through each image in the directory
for image_file in os.listdir(output_folder):
    if image_file.lower().endswith('.tif'):  # Check for both upper and lower case extensions
        print(f"\nProcessing image: {image_file}")
        image_file_path = os.path.join(output_folder, image_file)
        
        # Open the image
        with rasterio.open(image_file_path) as dataset:
            
            # Loop through each band and print pixel values
            for i in range(1, dataset.count + 1):  # Band indices are 1-based in rasterio
                band_data = dataset.read(i)
                print(f"\nBand {i} Pixel Values:")

                # Print pixel values for a small subset (e.g., top-left 5x5 pixels)
                subset = band_data[:5, :5]  # Adjust the slice as needed
                for row in range(subset.shape[0]):
                    for col in range(subset.shape[1]):
                        print(f"Pixel ({row}, {col}): {subset[row, col]}")
                
                print("-" * 50)  # Separator for readability
```

### automates the process

```python
import os
import numpy as np
import rasterio

# Define the paths to your folders
tif_folder = r'C:\Users\User\Documents\Python Scripts\Data\RGB_Test\3129AA'
npy_folder = r'C:\Users\User\Documents\Python Scripts\Data\RGB_Test\Indices_6'
output_folder = r'C:\Users\User\Documents\Python Scripts\Data\RGB_Test\npy_tiff'

# Loop through all the TIFF files in the folder
for tif_filename in os.listdir(tif_folder):
    if tif_filename.endswith('.tif'):
        # Full path to the TIFF file
        tif_path = os.path.join(tif_folder, tif_filename)

        # Extract metadata from the TIFF file
        with rasterio.open(tif_path) as src:
            transform = src.transform
            crs = src.crs
            metadata = src.meta.copy()

        # Construct the corresponding .npy filename
        base_name = os.path.splitext(tif_filename)[0] # [0] selects the basename and excludes the extension
        npy_filename = f'VDVI_WBI_UPDATED_VARI_UPDATED_NGRDI_UPDATED_IRG_UPDATED_updated_{base_name}.npy' #update with the prefix of all the files, ensure all files have the same prefix
        npy_path = os.path.join(npy_folder, npy_filename)

        # Ensure the corresponding .npy file exists
        if os.path.exists(npy_path):
            # Load the NumPy array
            array_data = np.load(npy_path)

            # Update the metadata for the output TIFF
            metadata.update({
                'driver': 'GTiff',
                'count': array_data.shape[0],  # Number of bands
                'height': array_data.shape[1],  # Rows
                'width': array_data.shape[2],  # Columns
                'dtype': 'float16',  # Data type
                'crs': crs,  # Coordinate Reference System
                'transform': transform  # Affine transformation
            })

            # Construct the output TIFF filename
            output_tif_filename = f'output_{base_name}.tif'
            output_tif_path = os.path.join(output_folder, output_tif_filename)

            # Write the NumPy array to a new TIFF file
            with rasterio.open(output_tif_path, 'w', **metadata) as dst:
                for band in range(array_data.shape[0]):
                    dst.write(array_data[band], band + 1)  # Write each band to the corresponding index
        else:
            print(f"Corresponding .npy file not found for {tif_filename}")
```
